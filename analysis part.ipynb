```python
def get_season(row):
    """
    This function takes a row from the DataFrame and returns the season based on the month.
    """
    month = row['Months']
    if month in [2, 3, 4, 5]:
        return 'Pre-Monsoon'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

# Add a 'season' column to your DataFrame using the apply function
df1['season'] = df1.apply(get_season, axis=1)
#creating the boxplot
sns.boxplot(
    x='season',
    y='PM2.5_mean',
    
    data= df1
)
plt.title('seasonwise effect of PM2.5 on monsoon')
plt.xlabel("Season")
plt.ylabel("Pm2.5_mean")
plt.show()
```
